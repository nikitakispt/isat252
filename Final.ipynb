# COVID-19 Data Analysis

COVID-19 is a respiratory virus that is part of a large family of viruses called Corona. COVID-19 was first detected in Wuhan, China in December of 2019. COVID-19 has now been detected in 70 locations internationally and has caused a global pandemic. 

Data was collected from European Centre for Disease Prevention and Control, an agency of the European Union. The data consists of COVID-19's geographical distribution world wide. 

Example #1
COVID-19 deaths per continent

```python
import pandas
df = pandas.read_excel ("s3://isat252nikita/COVID-19-geographic-disbtribution-worldwide.xlsx")

continents_name = df.groupby('continentExp').sum()['deaths']
continents_name.nlargest(6).plot.bar()
```

The data shows America with around 180k deaths, Europe with 175k deaths, Asia with 40k deaths, and Oceania with below 10k deaths. 

Example #2
COVID-19 cases per day of Cyprus

```python
cyprus_data = df.loc[df['countriesAndTerritories'] == 'Cyprus']

cyprus_data.plot(x='dateRep',y='cases')
```

In this line chart of Cyrpus, we see a spike in cases the 1st of April and a steady decline to the 1st of June with a few occasional increases. 

Example #3
Number of cases in relation to the number of deaths in Egypt

```python
egypt_data = df.loc[df['countriesAndTerritories'] == 'Egypt']

egypt_data.plot.scatter(x='cases',y='deaths')
```

As there are more cases there are more deaths. THe graph shows a positive relationship between X and Y. 

Example #4
The last 15 days of recorded data in Norway

```python
norway_data = df.loc[df['countriesAndTerritories'] == 'Norway']
norway_data[:15] #the last 15 days
```

Looking at the data of Norway's last recorded 15 days, there's no trends seen in cases nor deaths. Cases peaked on the last recorded day.
